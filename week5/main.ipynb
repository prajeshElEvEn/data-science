# Machine Learning Models


## Logistic Regression


### Importing Libraries

```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="white", color_codes=True)
```

### Importing Dataset

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df["Species"].value_counts()
```

### Scatter Plot


```python
sns.FacetGrid(df, hue="Species", height=6).map(
    plt.scatter, "Petal.Length", "Sepal.Width").add_legend()
```

### Converting categorical varibales into numbers


```python
flower_mapping = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
df["Species"] = df["Species"].map(flower_mapping)
```

```python
df.head()
```

### Preparing inputs and outputs


```python
X = df[['Sepal.Length', 'Sepal.Width', 'Petal.Length', 'Petal.Width']].values
y = df[['Species']].values
```

## Training Model


```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

### Model Accuracy


```python
model.score(X, y)
```

### Model Predictions


```python
expected = y
predicted = model.predict(X)
predicted
```

### Model Summary

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected, predicted))
```

```python
print(metrics.confusion_matrix(expected, predicted))
```
