# Data Analysis

```python
import pandas as pd
```

```python
data = {'Name': ['John', 'Anna', 'Peter', 'Linda', 'John'],
        'Age': [35, 28, None, 32, 35],
        'Salary': [2500, 3500, 4000, 4000, 2500]}
```

```python
df = pd.DataFrame(data)
```
